```python
from LDA_XGB.pipeline import CopathologyPipeline
from LDA_XGB.data_processor import *
from LDA_XGB.visualizer import CopathologyVisualizer
import pandas as pd
import os

path_all_dx_k_18 = 'C:/Users/WooSikKim/Desktop/Research/projects/co_pathology/scripts/stage_copath/STAGE_CoPathology/LDA_XGB/models/wsev_smc_all_dx_with_cn_k_18.pkl'
mdl_all_dx_k_18 = CopathologyPipeline.load(path_all_dx_k_18)
print(mdl_all_dx_k_18._region_cols)
```

```python
raw_df = pd.read_csv('C:/Users/WooSikKim/Desktop/Research/projects/co_pathology/scripts/stage_copath/data/resilience_data/20260219_MergedData_ver10.csv')
region_cols =  raw_df.loc[:,'VA/2_zscore':'VA/2035_zscore'].columns
# print(region_cols)
mdl_all_dx_k_18._region_cols = region_cols.to_list()
print(mdl_all_dx_k_18._region_cols)

print(raw_df.shape)
raw_df.dropna(subset=region_cols, inplace=True)
raw_df.drop_duplicates(subset='FULL_ID', inplace=True)
print(raw_df.shape)

lda_inference_df = mdl_all_dx_k_18.predict_new_subjects(inp_df=raw_df, subject_col='FULL_ID')
lda_inference_df = pd.merge(raw_df, lda_inference_df, left_on='FULL_ID', right_on='SUBJ_ID', how='left')
print(lda_inference_df.shape)
```
